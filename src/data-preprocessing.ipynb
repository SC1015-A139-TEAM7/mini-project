# Importing Necessary Libraries

```python
import requests
import numpy as np
import pandas as pd
from io import StringIO
```

# Data Extraction from Alphavantage API 

```python
time_length = 'TIME_SERIES_DAILY_ADJUSTED'
equity = 'SPY'
interval_length = 'Daily'
datatype = 'csv'
url = f'https://www.alphavantage.co/query?function={time_length}&symbol={equity}&interval={interval_length}&\
outputsize=full&apikey=LSY1P7SPJOP4UKMB&datatype={datatype}'
response = requests.get(url)
```

```python
data_string = StringIO(response.text)
stock_data = pd.read_csv(data_string)
stock_data
```

```python
stock_data.info()
```

# Data Cleaning

```python
stock_cleaned_data = pd.DataFrame(stock_data)
```

```python
stock_cleaned_data['timestamp'] = pd.to_datetime(stock_cleaned_data['timestamp'])
stock_cleaned_data.set_index('timestamp', drop = True, inplace = True)
stock_cleaned_data = stock_cleaned_data.rename_axis('date')
stock_cleaned_data = stock_cleaned_data.rename(columns= {'adjusted_close' : 'adjclose'})
```

```python
stock_cleaned_data.drop(['dividend_amount', 'split_coefficient'], axis=1, inplace=True)
stock_cleaned_data = stock_cleaned_data[::-1]
```
