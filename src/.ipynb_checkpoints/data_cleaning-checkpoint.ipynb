```python
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
time_length = 'TIME_SERIES_DAILY_ADJUSTED'
equity = 'SPY'
interval_length = 'Daily'
url = f'https://www.alphavantage.co/query?function={time_length}&symbol={equity}&interval={interval_length}&\
outputsize=full&apikey=LSY1P7SPJOP4UKMB'
r = requests.get(url)
data =  pd.DataFrame(r.json())
```

```python
data
```

```python
stock_data = pd.DataFrame(data)[6:].dropna(axis = 1)
```

```python
stock_data.head()
```

```python
dic = {'open' : [float(stock_data.loc[index, f'Time Series (Daily)'].get('1. open')) for index in stock_data.index],
      'high' : [float(stock_data.loc[index, f'Time Series (Daily)'].get('2. high')) for index in stock_data.index],
      'low' : [float(stock_data.loc[index, f'Time Series (Daily)'].get('3. low')) for index in stock_data.index],
      'close' : [float(stock_data.loc[index, f'Time Series (Daily)'].get('4. close')) for index in stock_data.index],
      'adjclose' : [float(stock_data.loc[index, f'Time Series (Daily)'].get('5. adjusted close')) for index in stock_data.index],
      'volume' : [int(stock_data.loc[index, f'Time Series (Daily)'].get('6. volume')) for index in stock_data.index]}
```

```python
stock_cleaned_data = pd.DataFrame(dic, index = stock_data.index)
```

```python
stock_cleaned_data.head()
```

```python
stock_cleaned_data['date'] = pd.to_datetime(stock_cleaned_data.index)
stock_cleaned_data.set_index('date', drop = True, inplace = True)
```
